```python
from bs4 import BeautifulSoup
import requests
from selenium import webdriver
from selenium.webdriver import Chrome
from time import sleep

from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
```

```python
def parse2(url):
    options = webdriver.ChromeOptions()
    options.add_argument('--ignore-certificate-errors')
    options.add_argument('--incognito')
    options.add_argument('--headless')
    driver = webdriver.Chrome(ChromeDriverManager().install(), options=options)
    driver.get(url)
    sleep(3)
    sourceCode = driver.page_source
    return sourceCode, driver
```

```python
first_half = 'https://www.salary.com/research/salary/alternate/ophthalmologist-salary/' 
second_half = '#:~:text=The%20average%20Ophthalmologist%20salary%20in,falls%20between%20%24281%2C100%20and%20%24382%2C500.'
```

```python
abreviations = ['al', 'ak', 'az', 'ar', 'ca', 'co', 'ct', 'de', 'fl', 'ga', 'hi', 'id', 'il', 'in', 'ia', 'ks', 'ky', 'la', 'me', 'md', 'ma', 'mi', 'mn', 'ms', 'mo', 'mt', 'ne', 'nv', 'nh', 'nj', 'nm', 'ny', 'nc', 'nd', 'oh', 'ok', 'or', 'pa', 'ri', 'sc', 'sd', 'tn', 'tx', 'ut', 'vt', 'va', 'wa', 'wv', 'wi', 'wy', 'dc']
```

```python
states = []
salaries = []
for x in abreviations:
    url = first_half + x + second_half
    sourceCode, driver = parse2(url)
    soup = BeautifulSoup(sourceCode)
    mydivs = soup.find_all("text", {"id": "top_salary_value"})
    mydivs = mydivs[0].find_all("tspan")
    salary = int(mydivs[0].text[1:].replace(",", ""))
    states.append(x.upper())
    salaries.append(salary)
    print(states)
    print(salaries)
```
